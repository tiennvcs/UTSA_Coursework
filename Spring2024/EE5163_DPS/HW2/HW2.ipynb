```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
n = np.linspace(0, 4*np.pi, 100)
```

```python
x1 = np.cos(n)
x2 = np.sin(n)
x3 = np.e**(1j*n)
```

```python
plt.plot(n, x1, label='cos(n)')
plt.plot(n, x2, label='sin(n)')
plt.legend(loc='best')
plt.show()
```

```python
x3_real = x3.real
x3_img = x3.imag

plt.plot(n, x3_real, label=r'Real($e^{jn}$)')
plt.plot(n, x3_img, label=r'Imagine($e^{jn}$)')
plt.legend(loc='best')
plt.show()
```
